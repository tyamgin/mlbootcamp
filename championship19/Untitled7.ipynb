```python
class KerasModel(MyModel):
    def fit(self, data):
        self.scaler = StandardScaler()
        X = self.get_X(data)
        self.scaler.fit(X)
        X = self.scaler.transform(X)
        y = data['label'].values
        
        model = keras.models.Sequential()
        self.model = model
        # Input - Layer
        model.add(keras.layers.Dense(10, activation = "relu", input_shape=(X.shape[1], )))
        # Hidden - Layers
        model.add(keras.layers.Dropout(0.3, noise_shape=None, seed=None))
        model.add(keras.layers.Dense(10, activation = "relu"))
        model.add(keras.layers.Dropout(0.2, noise_shape=None, seed=None))
        model.add(keras.layers.Dense(10, activation = "relu"))
        # Output- Layer
        model.add(keras.layers.Dense(1, activation = "sigmoid"))
        model.summary()
        # compiling the model
        model.compile(
            optimizer = "adam",
            loss = "binary_crossentropy",
            metrics = ["accuracy"]
        )
        results = model.fit(
            X, y,
            epochs=200,
            batch_size=1
        )

    def predict(self, X):
        X = self.scaler.transform(self.get_X(X))
        return self.model.predict(X)
        
        
        proba_result = self.model.predict_proba(X)
        return proba_result[:, 1]

```
