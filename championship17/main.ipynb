```python
def create_features(data):
    data = data.copy()
    
    data.drop(['audit_experiment', 'date'], 1, inplace=True)
    
    data['audit_old'] = data['audit_timestamp'] - data['metadata_createdAt']
    data['user_old'] = data['audit_timestamp'] - data['user_create_date'] / 1000
    data['membership_join_old'] = data['audit_timestamp'] - data['membership_joinDate'] / 1000
    
    if 'liked' in data.columns:
        data['label'] = data['liked']
        data.drop(['liked', 'clicked', 'viewed', 'disliked', 'reshared', 'ignored', 'commented', 'complaint', 
                   'unliked'], 1, inplace=True)

    
    data.drop(['membership_joinRequestDate', 'auditweights_feedOwner_RECOMMENDED_GROUP', 
               'auditweights_friendCommentFeeds', 'auditweights_friendCommenters', 'auditweights_hasDetectedText',
               'auditweights_hasText', 'auditweights_onlineVideo', 'auditweights_source_MOVIE_TOP',
               'auditweights_userOwner_CREATE_IMAGE', 'auditweights_userOwner_UNKNOWN',
               'auditweights_userOwner_VIDEO',
              
               'user_create_date', 'membership_joinDate',
               
              ], 1, inplace=True)
    
    # nan or single value
    for c in ('userOwnerCounters_USER_PROFILE_VIEW', 'auditweights_isRandom',
              'auditweights_notOriginalPhoto', 'auditweights_processedVideo'):
        data[c] = data[c].isna()
        
    data['membership_upd_old'] = data.audit_timestamp - data.membership_statusUpdateDate
    
    #date.drop(['auditweights_notOriginalPhoto', 'auditweights_processedVideo', 
    #           'auditweights_userOwner_CREATE_COMMENT', 'auditweights_userOwner_USER_FEED_REMOVE'], 1, inplace=True)

    return data
```

```python
train = create_features(train_data)
```

```python
class LgbModel(MyModel):
    def fit(self, data):
        X = self.get_X(data)
        self.columns = X.columns
        lgb_train = lgb.Dataset(X, data['label'].values)
        
        #if self.verbose >= 2:
        print('Starting train: %s' % datetime.datetime.now())
        params = self.params.copy()
        num_boost_round = params['num_boost_round']
        del params['num_boost_round']
        params['metric'] = 'auc'
        self.model = lgb.train(
            params,
            lgb_train,
            num_boost_round=num_boost_round
        )
    def predict(self, X):
        proba = self.model.predict(self.get_X(X))
        return proba
```
